Question 2: Open the attached file ``zapis_labview.lvm`` (it is a record from the LabView program) and read the header in the form of a dictionary ('field name': value).

```python
labview = open('zapis_labview.lvm', mode='r') # Open the file

data_list = []  # Prepare an empty list to which we will append values
for line in labview:
    item = line.split()  # Split the line into two columns
    if len(item) is 2:
         data_list.append((item[0], item[1]))  # Store pairs of tuples in the list, we will need them to build the dictionary
    else:
        break  # Stop once End of header is reached

labview.close() # Close the file

# Build the dictionary
dictionary = dict(data_list)

print(dictionary)

# Check whether it works
print(dictionary['Time'])
```

Question 3:  In the file from the previous point, read the data (use the ``replace()`` function to swap the decimal comma for a period and then convert the string to a number. Find the appropriate documentation/help).

```python
# Variant 1 (faster)

labview = open('zapis_labview.lvm', mode='r')

save = False
col_1 = []
col_2 = []
for line in labview:
    item = line.split()

    if 'Comment' in item:  # Start saving from the line after comment
        save = True
        continue

    if save is True:
        col_1.append(float(item[0].replace(',', '.')))
        col_2.append(float(item[1].replace(',', '.')))

print(col_1)
print(col_2)
```

```python
# Variant 2 (slower)

labview = open('zapis_labview.lvm', mode='r')

text = labview.read()  # Read the whole file, we get the text
# print(text)

# Replace the decimal comma with a decimal period
text = text.replace(',', '.')
#print(text)

data_list = text.split()  # Split the text into individual pieces, build a list

_ = data_list.index('Comment') # Find where the last word lies in the list and where the data begins

# Delete all elements that are not data
del data_list[:_+1]

# Convert to float and store in two columns
col_1 = []
col_2 = []

for number, element in enumerate(data_list):
    if number is 0:
        col_1.append(float(element))
    elif number%2 is 0:
        col_1.append(float(element))
    else:
        col_2.append(float(element))

print(col_1)
print(col_2)
```

Question 5: Write a function that receives a list.
* If the list contains words, then it returns their length.
* If the list contains numeric values, then it returns their values, increased by one.

```python
# Let's write the function
# Variant 1
def rearrange(data_list):
    if isinstance(data_list[0], str):    # [0] because data_list is a list, while the individual elements are strings
        result = [len(_) for _ in data_list]    # _ is a temporary variable, its name is irrelevant
        print(result)
    elif isinstance(data_list[0], (float, int)):  # A number can be an int or a float!
        result = [_ + 1 for _ in data_list]
        print(result)
    else:
        print('Your list is neither text nor numeric values. Please check!')

# Variant 2 (shorter)
def rearrange_2(data_list):
    result = [float(s)+1 if isinstance(s, (float, int)) else len(s) for s in data_list]
    print(result)

# Let's test with words
list_words = ['student', 'computer', 'house', 'table', 'Oops!']

rearrange(list_words)   # Call the function

rearrange_2(list_words)

# Let's test with numbers (float and int)
list_numbers1 = [3, 7, 1, 11, 9]
list_numbers2 = [4.2, 4.9, 11.3, 13.2]

rearrange(list_numbers1)

rearrange(list_numbers2)

rearrange_2(list_numbers2)
```

Question 6: Extend the function above: if an empty list (of length 0) is passed to the function, then it should raise an ``Exception`` with an appropriate description.


```python
# Let's extend the function
def rearrange(data_list):
    """ This function reads a list (which must not be empty).
 If it is a list of words, then it returns their length.
 If it is a list of numbers, then it adds the value 1 to them.
    """
    if len(data_list) is 0:
        raise Exception('The list is empty!')

    if isinstance(data_list[0], str):    # [0] because data_list is a list, while the individual elements are strings
        result = [len(_) for _ in data_list]    # _ is a temporary variable, its name is irrelevant
        print(result)
    elif isinstance(data_list[0], (float, int)):  # A number can be an int or a float!
        result = [_ + 1 for _ in data_list]
        print(result)
    else:
        print('Your list is neither text nor numeric values. Please check!')

# Let's test with an empty list
list_empty = []
rearrange(list_empty)

# Let's test with numbers   - we see that the code after the exception is not executed at all
rearrange(list_numbers1)
```

Question 7: Prepare a new function that will call the function from point 6 and catch exceptions.

```python
# New function
def rearrange_2(data_list):
    """ Does something similar to the rearrange function, but has a more robust way of handling errors"""
    try:

        rearrange(data_list) # If no error is raised, then we call the existing function

    except Exception as error:   # Print the error
        print(error)

# Let's test with an empty list
list_empty = []
rearrange_2(list_empty)

# Let's test with an ok list  - we see that in this case the code continues to execute
rearrange_2(list_numbers2)
```

Question 9: Demonstrate the use of arguments with default values.

```python
# Let's make a function that computes the volume of a cone

# We will need the number pi, so we import the numpy module (we import it outside the function, so that it is imported only once)
import numpy as np

def cone_volume(radius = 1, height = 1):
    """ The program computes the volume of a cone, the math module must be imported for it to work properly"""
    result = 1/3 * np.pi * radius**2 * height
    print('Data used: radius = {0:g} height ={1:g}'.format(radius, height))
    print('The volume of the cone is: {:5.3f}'.format(result))

# If we do not pass the radius or the height or both to the function, then the function uses the default values:
cone_volume()

cone_volume(3, )

cone_volume(height = 5)
```
