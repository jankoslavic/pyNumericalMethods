Question 5: Explain the difference between static and dynamic variable definition.

* A static variable is bound to both a type and an object at the same time (a = 9 and a = 'house' would raise an error).


* A dynamic variable is bound only to an object; the type can be changed freely. The example below works nicely because variables in Python are dynamic.

```python
a = 9
a
```

```python
a = 'house'
a
```

Question 10: Typical operations on lists.

You can find more about operations here: https://docs.python.org/3.4/tutorial/datastructures.html#more-on-lists

```python
list_3 = list_2.copy()  # We copy the list

list_3.append(18)  # We modify only the new list

list_3
```

```python
list_2  # We check whether we also modified the old list
```

```python
list_2.index('list')  # Let's see at which position in the list the word 'list' is located
```

```python
# Another example of using the counter
list_4 = [1, 2, 1, 3, 1, 4, 2, 2, 1]

# How many times the number 1 appears in list_4
list_4.count(1)
```

```python
# How many times the number 2 appears in list_4
list_4.count(2)
```

```python
# A list can also be rearranged into reverse order
list_4.reverse()
list_4
```

```python
# They can also be sorted
list_4.sort()
list_4
```

```python
# And a list of lists is in fact a matrix
list_5 = [3, 7, 7]

matrix = [list_5,
          [1, 1, 1],
          [3, 6, 9]]

matrix
```
