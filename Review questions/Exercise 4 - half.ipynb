Question 1: Using an example, explain *procedural* and *functional* programming.

```python
# Example of computing a factorial
n = 5

f = 1
for i in range(n):
    f *= i+1

result = f
print(result)
```

```python
from functools import reduce
#?reduce()
def factorial_2(n):
    return reduce(lambda x,y: x*y, range(1, n+1))

print(factorial_2(5))
```

Question 2: Define a simple object and write a few methods for that object.

```python
# Objects are based on classes
import numpy as np

class Cylinder:
    """ An object that stores or computes the properties of a cylinder."""

    def __init__(self, radius=2, height=5, density=0): # Object constructor
        self.radius = radius
        self.height = height
        self.density = density

    kind = 'Body'  # A property that will be the same for all objects

    def set_radius(self, radius):  # Change the radius
        self.radius = radius

    def set_height(self, height):  # Change the height
        self.height = height

    def set_density(self, density): # Change the density
        self.density = density

    def volume(self):  # Computes the volume from the current data
        _ = np.pi * self.radius**2 * self.height
        return _

    def mass(self):  # Computes the mass from the current volume
        _ = self.volume() * self.density  # self.volume() !! Otherwise it is method*int, which returns an error
        return _
```

```python
# Create an object
cylinder1 = Cylinder(3, 1, 1000)

# Print the volume and the mass
print('Volume: ', cylinder1.volume())
print('Mass: ', cylinder1.mass())

# change the radius and the density
cylinder1.set_radius(5)
cylinder1.set_density(2500)

# Print the new results
print('New volume: ', cylinder1.volume())
print('New mass:', cylinder1.mass())
```
