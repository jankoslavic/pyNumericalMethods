## The ``for`` and ``while`` loops:

When we want the code to run for every element of a list or a tuple:


```python
for element in my_tuple:
    print(2*element)
```

```python
my_tuple = (1, 2, 3, 4)
```

Let's run the same code as above once more. Will the result be the same?

```python
for element in my_tuple:
    print(2*element)
```

We often use the ``for`` loop together with the ``range`` function:

```python
for i in range(5):
    print(i)
```

```python
tuple(range(10))
```

When we want to run code *as long as* a certain condition holds:

```python
new_list = []
i = 0

while my_tuple[i] < 4:
    new_list.append(my_tuple[i])
    i += 1

print(tuple(new_list))
```

In the cell above we learned quite a few new things:
* We can add elements to a list with the ``list.append`` method.
* ``i += 1`` is a shorter equivalent of the code ``i = i + 1``.
* Using a ``for`` loop is a way to "change" the contents of an existing tuple.
* With a ``while`` loop we have to update the stopping condition ourselves!

## Defining your own functions

When we want to use some code more than once, it is good to define it as a function. The syntax is as follows:

```python
def print_all_doubled(items):
    for element in items:
        print(2*element)
```

```python
print_all_doubled(my_tuple)
```

We often want to document our functions for easier future use:

```python
def print_all_doubled_2(items):
    """
    A function that prints twice the value of each element in the given list.
    """
    for element in items:
        print(2*element)
```

```python
help(print_all_doubled_2)
```

Sometimes we expect our function to return a result that we can reuse:

```python
def print_all_doubled_3(items):
    """
    A function that prints twice the value of each element in the given list.
    Returns a list of the doubled elements of the given list.
    """
    doubled_list = []

    for element in items:
        doubled_list.append(2*element)
        print(2*element)

    return doubled_list
```

```python
doubled = print_all_doubled_3(my_tuple)
doubled
```

# Example: using Python to size a beam

We consider the beam in the figure.

<img src="img/konzola.png" style="width:550px">

## Problem

1. Compute the maximum bending stress in the beam $\sigma_{\text{max}}$.

* Compute and plot the deflection curve of the beam $y(x)$ for the given load $F$.


3. \* Define a function `bending_stress` that, for the given parameters, *plots* the deflection curve of the beam and *returns* the value of the maximum bending stress $\sigma_{\text{max}}$.

4. \* Among the various possible cross-section `heights` $h$, choose the smallest one for which the maximum bending stress in the beam is less than the allowable stress ($\sigma_{\text{max}} \leq \sigma_{\text{dop}}$).

## Equations


Section modulus and second moment of area of the cross-section:

$$W_z = \frac{a\,h^2}{6}$$

$$I = \frac{a\,h^3}{12}$$


Maximum bending moment (at point $A$):

$$M_{\text{max}} = F\, l$$


Maximum bending stress:

$$\sigma_{\text{max}} = \frac{M_{\text{max}}} {W_z}$$


Maximum deflection (at location $B$):

$$f = \frac{F\, l^3}{3\,E\,I}$$


Equation of the deflection curve:

$$u = l - \frac{x}{l}$$

$$y = \frac{f}{2} \, (2 - 3u + u^3)$$

## Data:

```python
import numpy as np
```

```python
a = 10      # mm
h = 15      # mm
F = 100     # N
l = 750     # mm
E = 210000  # MPa
s_dop = 240 # MPa

# question 4.
heights = np.linspace(1, 20, 39)
```

## Solution

### Maximum bending stress in the beam

```python
W_z = a*h**2/6
I_z = a*h**3/12
M_max = F*l
s_max = M_max / W_z
print('Maximum stress in the cross-section:', s_max, 'MPa')
```

### Deflection curve

First we prepare an array of the $x$ coordinate where we want to compute the deflection curve:

```python
x = np.linspace(0, l, 500) # we compute at 500 points along the beam
```

We compute the deflection curve $y(x)$:

Maximum deflection (at location $B$):

$$f = \frac{F\, l^3}{3\,E\,I}$$


Equation of the deflection curve:

$$u = 1 - \frac{x}{l}$$

$$y = \frac{f}{2} \, (2 - 3u + u^3)$$

```python
f = F*l**3 / (3*E*I_z)
u = 1 - x/l
y = f/2 * (2 - 3*u + u**3)
```

```python
np.max(y)
```

Plotting with `matplotlib.pyplot.plot`

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, -y)
```

Let's dress up the plot:

```python
plt.plot(x, -y, 'r', lw=3)
plt.plot([0, l], [0, 0], 'k', lw=3)
plt.axis('equal')
plt.xlabel('x [mm]')
plt.ylabel('y [mm]')
```

### The `bending_stress` function

3. \* Define a function `bending_stress` that, for the given parameters, *plots* the deflection curve of the beam and *returns* the value of the maximum bending stress $\sigma_{\text{max}}$.

```python
def bending_stress(a, h, F, l, E, do_plot=True):
    """
    For a cantilever beam and the given parameters, plots the deflection curve
    and returns the value of the maximum stress.
    """

    W_z = a*h**2/6
    I_z = a*h**3/12
    M_max = F*l
    s_max = M_max / W_z

    x = np.linspace(0, l, 500)
    f = F*l**3 / (3*E*I_z)
    u = 1 - x/l
    y = f/2 * (2 - 3*u + u**3)

    if do_plot:
        plt.plot(x, -y, 'r', lw=3)
        plt.plot([0, l], [0, 0], 'k', lw=3)
        plt.axis('equal')
        plt.xlabel('x [mm]')
        plt.ylabel('y [mm]')

    return s_max
```

```python
bending_stress(a, h, F, l, E)
```

### Sizing the beam


4. \* Among the various possible cross-section `heights` $h$, choose the smallest one for which the maximum bending stress in the beam is less than the allowable stress ($\sigma_{\text{max}} \leq \sigma_{\text{dop}}$).

```python
for height in heights:
    stress = bending_stress(a, height, F, l, E, do_plot=False)
    if stress < s_dop:
        print('Selected height:', height, 'mm')
        bending_stress(a, height, F, l, E, do_plot=True)
        break
```
