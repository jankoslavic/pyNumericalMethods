# Additional material

Let's try to write a simple program for integer division. The requirements are:

* Ask the user for two pieces of data, the value of the divisor and the dividend. Use the function `input(<message>)`!
* Convert both values from strings to floating-point numbers.
* Check whether the divisor is different from zero. If it is not, warn the user and stop.
* If we passed the test above, print the result of the integer division and the value of the remainder.

````python
result = input('Test input >> ')
type(result)
````

Solution:

````python
dividend = input('Dividend:')
divisor = input('Divisor:')
````

````python
float_dividend = float(dividend)
float_divisor = float(divisor)
float_divisor
````

````python
if float_divisor != 0:
    result = float_dividend // float_divisor
    remainder = float_dividend % float_divisor
    print(result, remainder)
else:
    print('Warning! The divisor is equal to 0')
````
